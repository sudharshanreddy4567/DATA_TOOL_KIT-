```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import Image
```

sin and cos graph

```python
x=np.linspace(0,3*np.pi,100)
sin_y=np.sin(x)
print(sin_y)

plt.plot(x,sin_y)
plt.show
plt.title("sign wave")
```

```python
h=np.linspace(0,3*np.pi)
s=np.sin(h)
print(s)
plt.plot(h,s)
plt.title("sine graph")
```

```python
a=np.linspace(0,2*np.pi,20)
b=np.cos(a)
c=np.sin(a)
plt.plot(a,b, '*')
plt.plot(a,c,'-')

plt.legend(loc='upper right')#right, lower right,center upeer right, lower left
plt.title("cos and sin wave")
plt.show()
```

```python
a=np.linspace(0,2*np.pi,20)
b=np.cos(a)
c=np.sin(a)
fig=plt.figure(figsize=(18,5))
plt.style.use('_mpl-gallery')
plt.plot(a,b, label='sine')
plt.plot(a,c,label='cose')

plt.legend(loc='upper right')#right, lower right,center upeer right, lower left
plt.title("cos and sin wave")
plt.show()
fig.savefig('sine_and_cosine.png')
```

```python
a=np.linspace(0,2*np.pi,100)
b=np.cos(a)
c=np.sin(a)

plt.figure(figsize=(18,5))
plt.style.use('Solarize_Light2')
# plt.plot(a,b,label='sine wave')
plt.plot(a,c,label='cose wave')
plt.title("sine ans cos grapg")
plt.show()

```

```python
n=np.linspace(0,4*np.pi,30)
n
h=np.sin(n)
j=np.cos(n)

fig=plt.figure()
plt.plot(n,h,'g',label='sine wave',color='red',linewidth=5)
plt.plot(n,j,'*',label='cose wave',color='pink',linewidth=5)
plt.style.use('seaborn-v0_8-colorblind')
plt.title('sine ans cose wave')
plt.legend(loc='upper right')
plt.savefig('sine ans cos2.png')

```

```python
import numpy as np
a=np.linspace(0,200,100)
b=np.linspace(0,150,100)

plt.plot(a,a,linestyle='solid',label='solid',color='pink',linewidth=4)

plt.plot(b,b+100,linestyle='solid',label='solid')
plt.plot(b,b+200,linestyle='dashdot',label='dashdot')
plt.plot(b,b-300,linestyle='solid',label='solid')
plt.style.use('fast')
plt.xlim([0,150])
plt.ylim([0,300])
plt.legend()
```

```python
n=np.linspace(0,4*np.pi,100)
k=np.sin(n)
k=plt.figure(figsize=(18,5))
plt.plot(n,k)
plt.title("Sine wave")

plt.xlabel("x-axis")
plt.ylabel("ylabel")
plt.legend(loc='upper right')
plt.savefig('iop')


```

```python
x=np.linspace(0,4*np.pi,100)
g=np.sin(x)
fig=plt.axes()
fig.plot(x,g)
plt.style.use('grayscale')
fig.set(xlim=(0,10),ylim=(-2,2),xlabel="it x",ylabel="its y",title="sine wave")
```

```python
n=5000
x=np.random.randn(n)
y=np.random.randn(n)
x1=np.random.randn(n)+2
y1=np.random.randn(n)+2

plt.scatter(x,y,color='red',alpha=0.3,s=4,label="group1")
plt.scatter(x1,y1,c="green",alpha=0.3,s=4,label="group2")
plt.legend(loc="upper right")
plt.title("scattering plot")
```
